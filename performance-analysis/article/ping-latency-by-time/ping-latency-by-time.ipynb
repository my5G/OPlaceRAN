```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Pings.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
dataframe.plot(kind='line', y='Média de Latency (ms)', label='Average Latency (ms)', ax=ax)
plt.fill_between(dataframe.index,dataframe['Média de Latency (ms)']+dataframe['DesvPad de Latency (ms)'],dataframe['Média de Latency (ms)']-dataframe['DesvPad de Latency (ms)'],alpha=0.2,label='Stdev')
plt.xlabel("Time (s)")
plt.ylabel("Latency (ms)")
plt.legend()
plt.savefig('out/Latency_by_time.pdf', bbox_inches='tight' )
plt.savefig('out/Latency_by_time.png',dpi=300, bbox_inches='tight' )

```
