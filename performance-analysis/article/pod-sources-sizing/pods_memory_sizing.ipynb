```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Pods-sizing.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
dataframe.plot(kind='line', y='cu Memory(Mi)', label='CU', ax=ax)
#ax.axhline(y=dataframe['cu CPU(m)'].max(),label='CU CPU(m) Average',linestyle = 'dotted')

dataframe.plot(kind='line', y='du Memory(Mi)', label='DU', ax=ax)
#ax.axhline(y=dataframe['du CPU(m)'].max(),label='DU CPU(m) Average',linestyle = 'dotted')

dataframe.plot(kind='line', y='ru Memory(Mi)', label='RU', ax=ax)
#ax.axhline(y=dataframe['ru CPU(m)'].max(),label='RU CPU(m) Average',linestyle = 'dotted')

plt.xlabel("Time (s)")
plt.ylabel("Memory(Mi)")
#plt.legend()
plt.savefig('out/Pods_MEM_Sizing_Millicores.pdf', bbox_inches='tight')
plt.savefig('out/Pods_MEM_Sizing_Millicores.png', dpi=300, bbox_inches='tight')

```
