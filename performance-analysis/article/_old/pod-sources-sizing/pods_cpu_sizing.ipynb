```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Pods-sizing.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
dataframe.plot(kind='line', y='cu CPU(m)', label='CU', ax=ax)
#ax.axhline(y=dataframe['cu CPU(m)'].max(),label='CU PU(m) Average',linestyle = 'dotted')
dataframe.plot(kind='line', y='du CPU(m)', label='DU', ax=ax)
#ax.axhline(y=dataframe['du CPU(m)'].max(),label='DU PU(m) Average',linestyle = 'dotted')
dataframe.plot(kind='line', y='ru CPU(m)', label='RU', ax=ax)
#ax.axhline(y=dataframe['ru CPU(m)'].max(),label='RU CPU(m) Average',linestyle = 'dotted')
plt.xlabel("Time (s)")
plt.ylabel("CPU (millicores)")
#plt.legend()
plt.savefig('out/Pods_CPU_Sizing_Millicores.pdf', bbox_inches='tight')
plt.savefig('out/Pods_CPU_Sizing_Millicores.png', dpi=300, bbox_inches='tight')
```
