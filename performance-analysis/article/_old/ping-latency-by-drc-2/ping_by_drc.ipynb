```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Ping_by_DRC_2.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
dataframe.plot(kind='line', y='DRC 5 Média de Latency(ms)', label='DRC 5', ax=ax, color='tab:blue')
plt.fill_between(dataframe.index,dataframe['DRC 5 Média de Latency(ms)']+dataframe['DRC 5 DesvPad de Latency(ms)']/2,dataframe['DRC 5 Média de Latency(ms)']-dataframe['DRC 5 DesvPad de Latency(ms)']/2,alpha=0.2, color='tab:blue')
plt.axhline(y=dataframe['DRC 5 Média de Latency(ms)'].mean(),linestyle='dotted',color='tab:blue', label='DRC 5 Average Latency=' + str(dataframe['DRC 5 Média de Latency(ms)'].mean()))

dataframe.plot(kind='line', y='DRC 7 Média de Latency(ms)', label='DRC 7', ax=ax, color='tab:green')
#plt.fill_between(dataframe.index,dataframe['DRC 5 Média de Latency(ms)']+dataframe['DRC 5 DesvPad de Latency(ms)']/2,dataframe['DRC 10 Média de Latency(ms)']-dataframe['DRC 10 DesvPad de Latency(ms)']/2,alpha=0.2,color='tab:orange')
plt.axhline(y=dataframe['DRC 7 Média de Latency(ms)'].mean(),linestyle='dotted',color='tab:green', label='DRC 7 Average Latency =' + str(dataframe['DRC 7 Média de Latency(ms)'].mean()))

dataframe.plot(kind='line', y='DRC 8 Média de Latency(ms)', label='DRC 8', ax=ax, color='tab:red')
plt.fill_between(dataframe.index,dataframe['DRC 5 Média de Latency(ms)']+dataframe['DRC 5 DesvPad de Latency(ms)']/2,dataframe['DRC 8 Média de Latency(ms)']-dataframe['DRC 8 DesvPad de Latency(ms)']/2,alpha=0.2,color='tab:red')
plt.axhline(y=dataframe['DRC 8 Média de Latency(ms)'].mean(),linestyle='dotted',color='tab:red', label='DRC 8 Average Latency =' + str(dataframe['DRC 8 Média de Latency(ms)'].mean()))

dataframe.plot(kind='line', y='DRC 10 Média de Latency(ms)', label='DRC 10', ax=ax, color='tab:orange')
plt.fill_between(dataframe.index,dataframe['DRC 5 Média de Latency(ms)']+dataframe['DRC 5 DesvPad de Latency(ms)']/2,dataframe['DRC 10 Média de Latency(ms)']-dataframe['DRC 10 DesvPad de Latency(ms)']/2,alpha=0.2,color='tab:orange')
plt.axhline(y=dataframe['DRC 10 Média de Latency(ms)'].mean(),linestyle='dotted',color='tab:orange', label='DRC 10 Average Latency =' + str(dataframe['DRC 10 Média de Latency(ms)'].mean()))


plt.xlabel("Time (s)")
plt.ylabel("Latency (ms)")
plt.legend()
plt.savefig('out/Ping_by_DRC.pdf', bbox_inches='tight')
plt.savefig('out/Ping_by_DRC.png', dpi=300, bbox_inches='tight')

```
