```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Decentralized_Nodes.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
dataframe.plot(kind='bar', y='Mdia de Memory(MiBytes)', yerr=[dataframe['DesvPad de Memory(MiBytes)']/2,dataframe['DesvPad de Memory(MiBytes)']/2],capsize=3,ax=ax)

plt.xlabel("Nodes ")
plt.ylabel("Memory (MiBytes)")
plt.legend()
plt.savefig('out/Nodes_MEM.pdf', bbox_inches='tight')
plt.savefig('out/Nodes_MEM.png', dpi=450, bbox_inches='tight')
```
