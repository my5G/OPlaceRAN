```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Pods-sizing-2.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
dataframe.plot(kind='line', y='Node MEMORY(MiBytes)', label='Node', ax=ax)
#dataframe.plot(kind='line', y='K8S CPU(Millicores)', label='K8S', ax=ax)
#dataframe.plot(kind='line', y='OAI Memory(MiBytes)', label='OAI', ax=ax)
plt.fill_between(dataframe.index,0,dataframe['K8S Memory(MiBytes)'],alpha=1, color='tab:red',label='K8S')
plt.fill_between(dataframe.index,dataframe['K8S Memory(MiBytes)'],dataframe['K8S Memory(MiBytes)']+dataframe['OAI Memory(MiBytes)'],alpha=0.3,color='tab:blue',label='OAI')

#plt.fill_between(dataframe.index,dataframe['K8S Memory(MiBytes)'],dataframe['Node CPU(millicores)']-dataframe['OAI Memory(MiBytes)'],alpha=0.3)
#plt.fill_between(dataframe.index,dataframe['Node CPU(millicores)'],dataframe['Node CPU(millicores)']+dataframe['K8S Memory(MiBytes)'],alpha=0.2, color='tab:blue')




plt.xlabel("Time (s)")
plt.ylabel("Memory (MiBytes)")
plt.legend()
plt.savefig('out/K8S_MEM_Overhead_MiBytes.pdf', bbox_inches='tight')
plt.savefig('out/K8S_MEM_Overhead_MiBytes.png', dpi=300, bbox_inches='tight')
```
