```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('PODs.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
plt.fill_between(dataframe.index,0,dataframe['K8S CPU (Millicores)'],alpha=0.3,label='K8S', color='tab:blue')

plt.fill_between(dataframe.index,dataframe['K8S CPU (Millicores)'],dataframe['K8S CPU (Millicores)']+dataframe['OPERATOR CPU (Millicores)'],\
    alpha=1,label='PlaceRAN Op.', color='tab:orange')

plt.fill_between(dataframe.index,dataframe['K8S CPU (Millicores)']+dataframe['OPERATOR CPU (Millicores)'],\
    dataframe['K8S CPU (Millicores)']+dataframe['OPERATOR CPU (Millicores)']+dataframe['CU CPU (Millicores)'],alpha=0.3,label='CU',color='tab:green') 

plt.fill_between(dataframe.index,dataframe['K8S CPU (Millicores)']+dataframe['OPERATOR CPU (Millicores)']+dataframe['CU CPU (Millicores)'],\
    dataframe['K8S CPU (Millicores)']+dataframe['OPERATOR CPU (Millicores)']+dataframe['CU CPU (Millicores)']+dataframe['DU CPU (Millicores)'],alpha=0.3,label='DU',color='tab:red') 

plt.fill_between(dataframe.index,dataframe['K8S CPU (Millicores)']+dataframe['OPERATOR CPU (Millicores)']+dataframe['CU CPU (Millicores)']+dataframe['DU CPU (Millicores)'],\
    dataframe['K8S CPU (Millicores)']+dataframe['OPERATOR CPU (Millicores)']+dataframe['CU CPU (Millicores)']+dataframe['DU CPU (Millicores)']+dataframe['RU CPU (Millicores)'],alpha=0.3,label='RU',color='tab:purple') 

ax.annotate('1', xy=(0,1072), xytext=(15, 5), textcoords='offset points', arrowprops=dict(arrowstyle='->', color='black'), fontsize=10)

plt.xlabel("Time (s)")
plt.ylabel("CPU (millicores)")
plt.legend()
plt.savefig('out/CPU_PODS.pdf', bbox_inches='tight')
plt.savefig('out/CPU_PODS.png', dpi=300, bbox_inches='tight')
```
