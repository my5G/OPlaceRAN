```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Nodes2.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
dataframe.plot( kind='bar', stacked=False)


plt.xlabel("Nodes")
plt.legend()
plt.savefig('out/nodes_resources.pdf', bbox_inches='tight')
plt.savefig('out/nodes_resources.png', dpi=300, bbox_inches='tight')
```
