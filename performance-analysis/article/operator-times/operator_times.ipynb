```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
names = ['Placement Start',\
     'Alg. Complete',\
          'Positioning start',\
               'CNFs positioned',\
                    'CNFs Loaded']
times = [0, 2, 31.58, 34, 70]
```

```python
# Choose some nice levels
levels = np.tile([-0.3, 0.3, -0.2, 0.2, -0.1, 0.1],
                 int(np.ceil(len(times)/6)))[:len(times)]

# Create figure and plot a stem plot with the date
fig, ax = plt.subplots(figsize=(8, 2),constrained_layout=True)
#ax.set(title="Operator initial operations time")

ax.vlines(times, 0, levels, color="tab:red")  # The vertical stems.
ax.plot(times, np.zeros_like(times), "-o",
        color="k", markerfacecolor="w")  # Baseline and markers on it.

# annotate lines
for d, l, r in zip(times, levels, names):
    ax.annotate(r, xy=(d, l),
                xytext=(-0.3, np.sign(l)*0.3), textcoords="offset points",
                horizontalalignment="center",
                verticalalignment="bottom" if l > 0 else "top")


# remove y axis and spines
ax.yaxis.set_visible(False)
plt.xlabel("Elapsed Time(s)")
#ax.margins(y=0.1)
plt.savefig('out/operator_times.pdf', bbox_inches='tight')
plt.savefig('out/operator_times.png', dpi=300, bbox_inches='tight')
```
