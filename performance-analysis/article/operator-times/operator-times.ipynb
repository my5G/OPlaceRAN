```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('operator-times.csv',  delimiter=';', header=0, index_col=0)
```

```python
del dataframe['Variation(%)']
```

```python
ax = dataframe.plot.bar()
ax.set_xlabel("Events")
ax.set_ylabel("Time(s)")
plt.savefig('out/operator-times.pdf', bbox_inches='tight')
plt.savefig('out/operator-times.png', dpi=300, bbox_inches='tight')
```
