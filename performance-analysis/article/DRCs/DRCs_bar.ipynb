```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('DRCs.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype('float16')
```

```python
from matplotlib.ticker import StrMethodFormatter

ax = dataframe.plot( kind='bar', stacked=False)
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}')) # No decimal places
plt.gca().yaxis.set_ticks([0,1,2])

plt.xlabel("DRCs")
plt.ylabel("Count")
plt.legend()
plt.savefig('out/DRCs_by_algorithm.pdf', bbox_inches='tight')
plt.savefig('out/DRCs_by_algorithm.png', dpi=300, bbox_inches='tight')
```
