```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('DRCs.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = dataframe.plot( kind='bar', stacked=False, color=['tab:red','tab:blue'])
plt.xlabel("DRC")
plt.ylabel("Count")
plt.legend()
ax.set_xticklabels(.astype(str), rotation='vertical')
plt.savefig('out/DRCs_by_algorithm.pdf', bbox_inches='tight')
plt.savefig('out/DRCs_by_algorithm.png', dpi=300, bbox_inches='tight')
```
