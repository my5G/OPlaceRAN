```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('DRCs.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = dataframe.plot( kind='pie',subplots=True,stacked=False,autopct='%1.1d%%',fontsize=8)
plt.legend()
plt.savefig('out/DRCs_by_algorithm_pie.pdf', bbox_inches='tight')
plt.savefig('out/DRCs_by_algorithm_pie.png', dpi=300, bbox_inches='tight')
```
