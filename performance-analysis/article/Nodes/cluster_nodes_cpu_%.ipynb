```python
import pandas as pd 
import matplotlib.pyplot as plt
import os

#os.chdir('performance-analysis/graphics/cluster_resources_time/')
dataframe = pd.read_csv('Nodes.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
print(dataframe)
```

```python
dataframe.plot(kind='line', y='Média de CPU',color='darkblue', label='Average')
plt.fill_between(dataframe.index,dataframe['Média de CPU']+dataframe['DesvPad de CPU'],dataframe['Média de CPU']-dataframe['DesvPad de CPU'],alpha=0.3,label='Stderr')
plt.xlabel("Time (s)")
plt.ylabel("CPU (%)")
plt.legend()
plt.savefig('out/Nodes_CPU_%.pdf')


```
