```python
import pandas as pd 
import matplotlib.pyplot as plt
import os

#os.chdir('performance-analysis/graphics/cluster_resources_time/')
dataframe = pd.read_csv('Nodes.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
print(dataframe)
```

```python
dataframe.plot(kind='line', y='Média de MEMORY',color='darkblue', label='Average')
plt.fill_between(dataframe.index,dataframe['Média de MEMORY']+dataframe['DesvPad de MEMORY'],dataframe['Média de MEMORY']-dataframe['DesvPad de MEMORY'],alpha=0.3, label='Stderr')
plt.xlabel("Time (s)")
plt.ylabel("Memory (%)")
plt.legend()
plt.savefig('out/Nodes_MEM_%.pdf')


```
