```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('PODs.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```
