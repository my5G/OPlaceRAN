```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('PODs.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
dataframe.plot(kind='line', y='CU CPU (Millicores)', label='CU', ax=ax)
dataframe.plot(kind='line', y='DU CPU (Millicores)', label='DU', ax=ax)
dataframe.plot(kind='line', y='RU CPU (Millicores)', label='RU', ax=ax)
#dataframe.plot(kind='line', y='K8S CPU (Millicores)', label='K8S', ax=ax)
dataframe.plot(kind='line', y='OPERATOR CPU (Millicores)', label='OP.', ax=ax)

ax.annotate('$\it{t0}$', xy=(0,0), xytext=(15, 15), textcoords='offset points', arrowprops=dict(arrowstyle='->', color='black'), fontsize=10)
plt.axvline(x = 0,linestyle ="dotted", color='black', alpha=0.3)

ax.annotate('$\it{t1}$', xy=(70,2002), xytext=(15, 15), textcoords='offset points', arrowprops=dict(arrowstyle='->', color='black'), fontsize=10)
plt.axvline(x = 70,linestyle ="dotted", color='black', alpha=0.3)

ax.annotate('$\it{t2}$', xy=(125,2002), xytext=(15, 15), textcoords='offset points', arrowprops=dict(arrowstyle='->', color='black'), fontsize=10)
plt.axvline(x = 125,linestyle ="dotted", color='black', alpha=0.3)

ax.annotate('$\it{t3}$', xy=(180,1639), xytext=(15, 15), textcoords='offset points', arrowprops=dict(arrowstyle='->', color='black'), fontsize=10)
plt.axvline(x = 180,linestyle ="dotted", color='black', alpha=0.3)

ax.annotate('$\it{t4}$', xy=(245,299), xytext=(15, 15), textcoords='offset points', arrowprops=dict(arrowstyle='->', color='black'), fontsize=10)
plt.axvline(x = 245,linestyle ="dotted", color='black', alpha=0.3)

ax.annotate('$\it{t5}$', xy=(300,1026), xytext=(15, 15), textcoords='offset points', arrowprops=dict(arrowstyle='->', color='black'), fontsize=10)
plt.axvline(x = 300,linestyle ="dotted", color='black', alpha=0.3)


plt.xlabel("Time (s)")
plt.ylabel("CPU (millicores)")
plt.legend()
plt.savefig('out/CPU_PODS.pdf', bbox_inches='tight')
plt.savefig('out/CPU_PODS.png', dpi=300, bbox_inches='tight')
```
