```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('PODS OAI vs K8S.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
plt.fill_between(dataframe.index,0,dataframe['K8S MEMORY (MiBytes)'],alpha=1, color='tab:red',label='K8S')
plt.fill_between(dataframe.index,dataframe['K8S MEMORY (MiBytes)'],dataframe['K8S MEMORY (MiBytes)']+dataframe['OAI MEMORY (MiBytes)'],alpha=0.3,color='tab:blue',label='OAI')


plt.xlabel("Time (s)")
plt.ylabel("MEMORY (MiBytes)")
plt.legend()
plt.savefig('out/MEM_PODS.pdf', bbox_inches='tight')
plt.savefig('out/MEM_PODS.png', dpi=300, bbox_inches='tight')
```
