```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Nodes4.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax=dataframe.plot( kind='bar',subplots=True)
ax[0].set_ylabel('Millicores')
ax[1].set_ylabel('MiBytes')
ax[0].set_title(None)
ax[1].set_title(None)
plt.savefig('out/nodes_resources_2g.pdf', bbox_inches='tight')
plt.savefig('out/nodes_resources_2g.png', dpi=300, bbox_inches='tight')
```
