```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")

dataframe = pd.read_csv('Nodes.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
dataframe.plot(kind='line', y='Mdia de CPU',color='darkblue', label='Average')
plt.fill_between(dataframe.index,dataframe['Mdia de CPU']+dataframe['DesvPad de CPU'],dataframe['Mdia de CPU']-dataframe['DesvPad de CPU'],alpha=0.3,label='Stdev')
plt.xlabel("Time (s)")
plt.ylabel("CPU (%)")
plt.legend()
plt.savefig('out/Nodes_CPU_%.pdf')
plt.savefig('out/Nodes_CPU_%.png')

```
