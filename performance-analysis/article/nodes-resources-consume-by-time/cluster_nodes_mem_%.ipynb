```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Nodes.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
dataframe.plot(kind='line', y='Mdia de MEMORY',color='darkblue', label='Average')
plt.fill_between(dataframe.index,dataframe['Mdia de MEMORY']+dataframe['DesvPad de MEMORY'],dataframe['Mdia de MEMORY']-dataframe['DesvPad de MEMORY'],alpha=0.3, label='Stdev')
plt.xlabel("Time (s)")
plt.ylabel("Memory (%)")
plt.legend()
plt.savefig('out/Nodes_MEM_%.pdf')
plt.savefig('out/Nodes_MEM_%.png')

```
