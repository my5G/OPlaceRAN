```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")

dataframe = pd.read_csv('Nodes.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
dataframe.plot(kind='line', y='Soma de MEMORY (bytes)', label='SUM Memory (bytes)',ax=ax)
dataframe.plot(kind='line', y='Soma de CPU (m)', label='SUM CPU (millicpu)',ax=ax)
plt.xlabel("Time (s)")
#plt.ylabel("Memory (%)")
plt.legend()
plt.savefig('out/Nodes_MEM_CPU.pdf')
plt.savefig('out/Nodes_MEM_CPU.png')
```
