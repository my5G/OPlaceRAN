```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")

dataframe = pd.read_csv('Nodes-resources-by-Nodes.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
dataframe.plot(kind='bar', y='Média de MEMORY%', label='Average MEM (%)', yerr=[dataframe['Média de MEMORY%']+dataframe['DesvPad de MEMORY%'],dataframe['Média de MEMORY%']+dataframe['DesvPad de MEMORY%']],capsize=3,ax=ax)
#dataframe.plot(kind='line', y='Máx. de CPU (m)',color='darkred',style='.', label='Max CPU (millicores)',ax=ax)
plt.xlabel("Nodes")
plt.ylabel("MEM (%)")
plt.legend()
plt.savefig('out/MEM_by_nodes.pdf')
plt.savefig('out/MEM_by_nodes.png')
```
