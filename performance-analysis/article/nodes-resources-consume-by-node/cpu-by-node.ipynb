```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")

dataframe = pd.read_csv('Nodes-resources-by-Nodes.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
dataframe.plot(kind='bar', y='Média de CPU%', label='Average CPU (%)', yerr=[dataframe['Média de CPU%']+dataframe['DesvPad de CPU%'],dataframe['Média de CPU%']+dataframe['DesvPad de CPU%']],capsize=3,ax=ax)
#dataframe.plot(kind='line', y='Máx. de CPU (m)',color='darkred',style='.', label='Max CPU (millicores)',ax=ax)
plt.xlabel("Nodes")
plt.ylabel("CPU (%)")
plt.legend()
plt.savefig('out/CPU_by_nodes.pdf',bbox_inches='tight')
plt.savefig('out/CPU_by_nodes.png', dpi=300, bbox_inches='tight' )
```
