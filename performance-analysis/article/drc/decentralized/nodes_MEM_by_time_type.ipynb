```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Decentralized_Nodes_Types.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
dataframe.plot(kind='line', y='1 Mdia de Memory(MiBytes)', label='Type 1 Average', ax=ax, color='tab:blue')
plt.fill_between(dataframe.index,dataframe['1 Mdia de Memory(MiBytes)']+dataframe['1 DesvPad de Memory(MiBytes)']/2,dataframe['1 Mdia de Memory(MiBytes)']-dataframe['1 DesvPad de Memory(MiBytes)']/2,alpha=0.2, color='tab:blue',label='Type 1 Stdev')

dataframe.plot(kind='line', y='2 Mdia de Memory(MiBytes)', label='Type 2 Average', ax=ax, color='tab:orange')
plt.fill_between(dataframe.index,dataframe['2 Mdia de Memory(MiBytes)']+dataframe['2 DesvPad de Memory(MiBytes)']/2,dataframe['2 Mdia de Memory(MiBytes)']-dataframe['2 DesvPad de Memory(MiBytes)']/2,alpha=0.2, color='tab:orange',label='Type 2 Stdev')

plt.xlabel("Time (s)")
plt.ylabel("Memory(MiBytes)")
plt.legend()
plt.savefig('out/Nodes_MEM_by_Time_Type.pdf', bbox_inches='tight')
plt.savefig('out/Nodes_MEM_by_Time_Type.png', dpi=300, bbox_inches='tight')
```
