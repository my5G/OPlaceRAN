```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Decentralized_Pods.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
dataframe.plot(kind='line', y='Mdia de Latency(ms)', label='DRC 8 Average', ax=ax, color='tab:blue')
plt.fill_between(dataframe.index,dataframe['Mdia de Latency(ms)']+dataframe['DesvPad de Latency(ms)']/2,dataframe['Mdia de Latency(ms)']-dataframe['DesvPad de Latency(ms)']/2,alpha=0.2, color='tab:blue',label='DRC 8 Stdev')
plt.axhline(y=dataframe['Mdia de Latency(ms)'].mean(),linestyle='dotted',color='tab:blue', label='DRC 8 Total Average')
plt.xlabel("Time (s)")
plt.ylabel("Latency (ms)")
plt.legend()
plt.savefig('out/Ping_by_DRC.pdf', bbox_inches='tight')
plt.savefig('out/Ping_by_DRC.png', dpi=300, bbox_inches='tight')
```
