```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")

dataframe = pd.read_csv('Pods.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
#print(dataframe)
```

```python
ax = plt.gca()
dataframe.plot(kind='line', y='cu Média de MEMORY(Mi)', label='CU Average Memory (Mi)', ax=ax)
plt.fill_between(dataframe.index,dataframe['cu Média de MEMORY(Mi)']+dataframe['cu DesvPad de MEMORY(Mi)'],dataframe['cu Média de MEMORY(Mi)']-dataframe['cu DesvPad de MEMORY(Mi)'],alpha=0.2,label='Stdev')
dataframe.plot(kind='line', y='du Média de MEMORY(Mi)',label='DU Average Memory (Mi)', ax=ax)
plt.fill_between(dataframe.index,dataframe['du Média de MEMORY(Mi)']+dataframe['du DesvPad de MEMORY(Mi)'],dataframe['du Média de MEMORY(Mi)']-dataframe['du DesvPad de MEMORY(Mi)'],alpha=0.2,label='Stdev')
dataframe.plot(kind='line', y='ru Média de MEMORY(Mi)', label='RU Average Memory (Mi)', ax=ax)
plt.fill_between(dataframe.index,dataframe['ru Média de MEMORY(Mi)']+dataframe['ru DesvPad de MEMORY(Mi)'],dataframe['ru Média de MEMORY(Mi)']-dataframe['ru DesvPad de MEMORY(Mi)'],alpha=0.2,label='Stdev')
plt.xlabel("Time (s)")
plt.ylabel("Memory (bytes)")
#plt.legend()
plt.savefig('out/Pods_MEM_bytes.pdf')
plt.savefig('out/Pods_MEM_bytes.png', dpi=300)

```
