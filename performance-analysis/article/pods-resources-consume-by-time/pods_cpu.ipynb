```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
dataframe = pd.read_csv('Pods.csv',  delimiter=';', header=0, index_col=0)
dataframe=dataframe.astype(float)
```

```python
ax = plt.gca()
dataframe.plot(kind='line', y='cu MÃ©dia de CPU(milicores)', label='CU Average CPU (millicores)', ax=ax)
plt.fill_between(dataframe.index,dataframe['cu MÃ©dia de CPU(milicores)']+dataframe['cu DesvPad de CPU(milicores)'],dataframe['cu MÃ©dia de CPU(milicores)']-dataframe['cu DesvPad de CPU(milicores)'],alpha=0.2,label='Stdev')
dataframe.plot(kind='line', y='du MÃ©dia de CPU(milicores)',label='DU Average CPU (millicores)', ax=ax)
plt.fill_between(dataframe.index,dataframe['du MÃ©dia de CPU(milicores)']+dataframe['du DesvPad de CPU(milicores)'],dataframe['du MÃ©dia de CPU(milicores)']-dataframe['du DesvPad de CPU(milicores)'],alpha=0.2,label='Stdev')
dataframe.plot(kind='line', y='ru MÃ©dia de CPU(milicores)', label='RU Average CPU (millicores)', ax=ax)
plt.fill_between(dataframe.index,dataframe['ru MÃ©dia de CPU(milicores)']+dataframe['ru DesvPad de CPU(milicores)'],dataframe['ru MÃ©dia de CPU(milicores)']-dataframe['ru DesvPad de CPU(milicores)'],alpha=0.2,label='Stdev')
plt.xlabel("Time (s)")
plt.ylabel("CPU (millicores)")
#plt.legend()
plt.savefig('out/Pods_CPU_Millicores.pdf')
plt.savefig('out/Pods_CPU_Millicores.png', dpi=300)

```
