```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sb
sb.set()
dataframe = pd.read_csv('Node.csv',  delimiter=';', header=0)
ax = sb.barplot(x='Node Number', y='Media', hue='Resource Type', data=dataframe)
for item in ax.get_xticklabels():
    item.set_rotation(45)
plt.savefig('out/Nodes_Resources_3.pdf')



```
